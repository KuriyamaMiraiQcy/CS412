```python
def bootstrap_data(training_data):
    resample_data = []
    for i in range(len(training_data)):
        resample_data.append(training_data[random.randrange(len(training_data))])
    return resample_data
```

```python
def lower_bound(result, val):
    start = 0
    end = len(result) - 1
    while start < end:
        mid = start + (end - start) // 2
        if result[mid] < val:
            start = mid + 1
        else:
            end = mid
    return start
```

```python
def upper_bound(result, val):
    start = 0
    end = len(result) - 1
    while start < end:
        mid = start + (end - start + 1) // 2
        if result[mid] > val:
            end = mid - 1
        else:
            start = mid
    return end
```

```python
print(lower_bound([2,2,3,5,6], 2))
upper_bound([2,2,3,5,6], 2)
```
